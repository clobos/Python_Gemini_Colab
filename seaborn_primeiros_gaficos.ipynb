```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

print("First 5 rows of the Iris dataset:")
display(iris.head())
```

### 1. Scatter Plot: Sepal Length vs. Sepal Width by Species

A scatter plot shows the relationship between two numerical variables. Here, we can see how sepal length and sepal width are distributed across different Iris species.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, s=100)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 2. Pair Plot: Relationships between all numerical variables

A pair plot creates a grid of scatter plots for each pair of variables, and histograms for each variable. This is useful for understanding the overall distribution and relationships within the dataset.

```python
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02) # Adjust title position
plt.show()
```

### 3. Histogram (Distribution Plot): Petal Length Distribution

A histogram shows the distribution of a single numerical variable. We can observe the frequency of different petal length values.

```python
plt.figure(figsize=(8, 6))
sns.histplot(iris['petal_length'], bins=10, kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### 4. Box Plot: Petal Width by Species

A box plot (or box-and-whisker plot) displays the distribution of a quantitative variable for different categories, showing the median, quartiles, and potential outliers. Here, we compare petal width across the three Iris species.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', hue='species', data=iris, palette='viridis', legend=False)
plt.title('Petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### 5. Violin Plot: Sepal Length by Species

A violin plot is similar to a box plot but also shows the probability density of the data at different values (like a kernel density estimate). It provides a richer view of the distribution shape.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_length', hue='species', data=iris, palette='muted', legend=False)
plt.title('Sepal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
