### Gráfico de Dispersão do Conjunto de Dados Iris

Este gráfico mostra a relação entre a largura (sepal_width) e o comprimento (sepal_length) da sépala para as diferentes espécies de Iris. Cada ponto representa uma flor, e as cores indicam a espécie (setosa, versicolor, virginica). As distribuições marginais (violino e rug) ajudam a entender a densidade dos dados para cada eixo.

```python
import plotly.express as px

# Load the Iris dataset
df_iris = px.data.iris()

# Create a scatter plot
fig = px.scatter(
    df_iris,
    x="sepal_width",
    y="sepal_length",
    color="species",
    marginal_y="violin",
    marginal_x="rug",
    hover_data=df_iris.columns
)

# Display the plot
fig.show()
```

### Box Plot do Comprimento da Sépala por Espécie do Conjunto de Dados Iris

Este box plot exibe a distribuição do comprimento da sépala (`sepal_length`) para cada uma das três espécies de Iris (setosa, versicolor, virginica). Cada caixa mostra o primeiro quartil (Q1), a mediana (Q2) e o terceiro quartil (Q3), enquanto os 'bigodes' (whiskers) indicam a variabilidade fora desses quartis. Pontos individuais fora dos 'bigodes' são considerados valores atípicos (outliers).

```python
import plotly.express as px

# Create a box plot for 'sepal_length' by 'species'
fig_box = px.box(df_iris, x="species", y="sepal_length", title="Box Plot do Comprimento da Sépala por Espécie")

# Display the plot
fig_box.show()
```

### Histograma do Comprimento da Sépala do Conjunto de Dados Iris

Este histograma mostra a distribuição do comprimento da sépala (`sepal_length`) para as diferentes espécies de Iris. É possível observar como o comprimento da sépala varia dentro de cada espécie e a sobreposição entre elas.

```python
import plotly.express as px

# Create a histogram for 'sepal_length'
fig_hist = px.histogram(df_iris, x="sepal_length", color="species", marginal="rug", title="Distribuição do Comprimento da Sépala por Espécie")

# Display the plot
fig_hist.show()
```
