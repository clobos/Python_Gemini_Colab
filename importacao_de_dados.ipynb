```python
import pandas as pd

# Path to the Excel file
excel_file_path = '/content/dados_colab.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path)

# Display the first 5 rows of the DataFrame
#display(df.head())
```

```python
# Create a contingency table between 'curso' and 'desempenho_geral'
contingency_table = pd.crosstab(df['curso'], df['desempenho_geral'])

# Display the contingency table
display(contingency_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the stacked bar chart
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Dist. de Desempenho G. por Curso')
plt.xlabel('Curso')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.legend(title='D. Geral', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
